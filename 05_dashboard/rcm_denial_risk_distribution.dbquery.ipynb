```sql
SELECT
    denial_risk_label,
    provider_state,
    COUNT(*)                                           AS claim_count,
    ROUND(AVG(denial_risk_score), 4)                   AS avg_risk_score,
    ROUND(
        COUNT(*) / SUM(COUNT(*)) OVER () * 100
    , 2)                                               AS pct_of_total
FROM rcm_platform.rcm_gold.denial_risk_scores
GROUP BY denial_risk_label, provider_state
ORDER BY avg_risk_score DESC
```
