```sql
SELECT
    rcm_health_grade,
    COUNT(*)                                           AS hospital_count,
    ROUND(AVG(rcm_health_score), 1)                    AS avg_score,
    ROUND(AVG(avg_ctp_ratio), 2)                       AS avg_ctp_ratio,
    ROUND(AVG(avg_revenue_gap_pct), 1)                 AS avg_revenue_gap_pct,
    ROUND(AVG(underpayment_rate_pct), 1)               AS avg_underpayment_rate,
    SUM(total_discharges)                              AS total_discharges
FROM rcm_platform.rcm_gold.hospital_scorecard
GROUP BY rcm_health_grade
ORDER BY avg_score DESC
```
