```sql
SELECT
    COUNT(DISTINCT provider_id)                        AS total_hospitals,
    COUNT(DISTINCT provider_state)                     AS states_covered,
    COUNT(DISTINCT drg_code)                           AS unique_drgs,
    SUM(total_discharges)                              AS total_discharges,
    ROUND(AVG(charge_to_payment_ratio), 2)             AS national_avg_ctp_ratio,
    ROUND(AVG(revenue_gap_pct), 2)                     AS national_avg_revenue_gap_pct,
    ROUND(SUM(total_revenue_gap) / 1e9, 2)             AS total_revenue_gap_billions,
    SUM(underpayment_flag)                             AS total_underpayment_flags,
    ROUND(SUM(underpayment_flag) / COUNT(*) * 100, 2)  AS national_underpayment_rate_pct
FROM rcm_platform.rcm_gold.fact_claims
```
