```sql
SELECT
    ar_aging_bucket,
    SUM(claim_count)                                   AS total_claims,
    SUM(total_discharges)                              AS total_discharges,
    ROUND(SUM(revenue_at_risk) / 1e9, 2)               AS revenue_at_risk_billions,
    ROUND(AVG(avg_days_in_ar), 0)                      AS avg_days_in_ar,
    ROUND(
        SUM(revenue_at_risk) /
        SUM(SUM(revenue_at_risk)) OVER () * 100
    , 2)                                               AS pct_of_total_ar
FROM rcm_platform.rcm_gold.ar_aging_buckets
GROUP BY ar_aging_bucket
ORDER BY avg_days_in_ar
```
